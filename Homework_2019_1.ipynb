```python
from mpl_toolkits.mplot3d import Axes3D

def two_d_gaussian(x, y, mu, cov):
    # only if x and y are independent
    denom = 1/2 * 1/np.pi * 1/(np.sqrt(cov[0][0]) * np.sqrt(cov[1][1]))
    mole = np.exp(-1/2 * (np.power(x - mu[0], 2)/np.sqrt(cov[0][0]) + (np.power(y - mu[1], 2)/np.sqrt(cov[1][1]))))
    z = mole/denom
    return z

def plot_twoDGaussian():
    # Setting parameters
    mu = np.asarray([0.0, 0.0])
    sigma = np.asarray([[1.0, 0.0], [0.0, 1.0]])

    # Generating coordinates and gaussian distribution
    x = np.linspace(-1, 1, 100)
    y = np.linspace(-1, 1, 100)
    x, y = np.meshgrid(x, y)
    pos = np.empty(x.shape + (2,))
    pos[:, :, 0] = x
    pos[:, :, 1] = y
    z = two_d_gaussian(x, y, mu, sigma)

    # Plot the graph
    fig = plt.figure(figsize=(13, 7))
    ax = plt.axes(projection='3d')
    ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='rainbow', alpha=0.9)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.savefig("twoDGaussian.png")

plot_twoDGaussian()
```
